```python
import cv2
import numpy as np
```

```python
# image= cv2.imread("https://assets.bwbx.io/images/users/iqjWHBFdfxIU/iKIWgaiJUtss/v2/1800x1350.jpg")

image= cv2.imread("/content/penguin.jpg")
```

```python
if image is not None:
  print("Image Loaded Successfully!")
  print("Imgae Shape :", image.shape)

else:
  print("Error Loading Image")
```

```python
from google.colab.patches import cv2_imshow

if image is not None:
  cv2_imshow(image)

else:
  print("Error Loading Image")
```

```python
# Shape of Image
original_height, original_width= image.shape[:2]
original_height, original_width
```

```python
# Gray Scale of Image
gray_image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
# Reshape the image
reshaped_image= gray_image.reshape(-1)
```

```python
original_shaped_image= gray_image.reshape(original_height, original_width)
```

```python
cv2_imshow(original_shaped_image)
```

```python
# Resize the Image
new_width, new_height= 200, 100
resized_img= cv2.resize(gray_image, (new_width, new_height))
cv2_imshow(resized_img)
```
